```python
#importing all the essential libraries for the project
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
%matplotlib inline
```

loading the dataset

```python
#load the dataset
df = pd.read_csv('/content/IRIS.csv')
```

```python
# shwowing the 1st 5 lines of the content table
df.head()
```

visualization of our dataset


```python
df.describe()
```

```python
#visualize the whole dataset
import seaborn as sns
import matplotlib.pyplot as plt

# Use white background
sns.set(style="white")

# Create pairplot
sns.pairplot(df, hue="species", palette="Set2")

# Show the plot
plt.show()
```

separating input columns and the output columns

```python
#separate features and target
data = df.values
X = data[:,0:4]
Y = data[:,4]
print(X)
```

```python
print(Y)
```

splitting the data into traning and testing

```python
#split the data to train and test datasheet
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)
print(X_test)
```

logistic regression model

```python
#logistic regression
from sklearn.linear_model import LogisticRegression
model_LR= LogisticRegression()
model_LR.fit(X_train, Y_train)
```

predicting


```python
prediction = model_LR.predict(X_test)
#calculating the accuracy
from sklearn.metrics import accuracy_score
print(accuracy_score(Y_test, prediction)*100)
```

```python
#detailed classification report
from sklearn.metrics import classification_report
print(classification_report(Y_test, prediction))
```
